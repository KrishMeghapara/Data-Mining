### 1)  First, you need to read the titanic dataset from local disk and display first five records



```python
import pandas as pd 
df = pd.read_csv("titanic.csv");
df.head(5)
```

##### 2)  Identify Nominal, Ordinal, Binary and Numeric attributes from data sets and display all values.



```python

nominal = ['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked']
for col in nominal:
    print(f"- {col}:\n{df[col].unique()}")

ordinal = ['Pclass']
for col in ordinal:
    print(f"- {col}:\n{sorted(df[col].unique())}")

binary = ['Survived']
for col in binary:
    print(f"- {col}:\n{df[col].unique()}")

numeric = ['PassengerId', 'Age', 'SibSp', 'Parch', 'Fare']
for col in numeric:
    print(f"- {col}:\n{df[col].tolist()}") 
```

### 3)  Identify symmetric and asymmetric binary attributes from data sets and display all values.



```python
binary_attrs = ['Survived']

print(" Asymmetric binary attributes:")
for attr in binary_attrs:
    print(f"- {attr}:\n{df[attr].unique()}")

print("\n Symmetric binary attributes:")
print("- None")
```

### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
arr = ["PassengerId","Survived","Pclass","Age","SibSp","Parch","Fare"];

for i in arr:
    print(i)
    print(f"\t mean : {df[i].mean()}")
    print(f"\t Standard deviation : {df[i].std()}")
    print(f"\t Min : {df[i].min()}")
    print(f"\t Max : {df[i].max()}")
    print(f"\t mode : {df[i].mode()[0]}") 
    print(f"\t Range : {df[i].max() - df[i].min() } ")
```

### 6)  For the qualitative attribute (class), count the frequency for each of its distinct values.



```python
df["Pclass"].value_counts()
```

### 8)  For multivariate statistics, you can compute the covariance and correlation between pairs of attributes.



```python
df.cov(numeric_only=True)
```

```python
df.corr(numeric_only=True)
```

### 9) Display the histogram for Age attribute by discretizing it into 8 separate bins and counting the frequency for each bin.

```python
import matplotlib.pyplot as plt
ax = df['Age'].hist(bins=8)

ax.set_title('Distribution of Age in 8 Bins')
ax.set_xlabel('Age')
ax.set_ylabel('Frequency')

plt.show()
```

### 10) A boxplot can also be used to show the distribution of values for each attribute.

```python
ax = df.boxplot(column=['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare'], vert=True)

ax.set_title('Boxplot of Numerical Attributes')
ax.set_xlabel('Values')

plt.show()
```

### 11)  Display scatter plot for any 5 pair of attributes , we can use a scatter plot to visualize their joint distribution.

```python
pair_indices = [
    ('Age', 'Fare'),
    ('SibSp', 'Parch'),
    ('Fare', 'Pclass'),
    ('Age', 'Survived'),
    ('Parch', 'Survived')
]

fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs = axs.flatten()

for i, (x, y) in enumerate(pair_indices):
    axs[i].scatter(df[x], df[y], color='skyblue', edgecolor='black')
    axs[i].set_xlabel(x)
    axs[i].set_ylabel(y)
    axs[i].set_title(f'{x} vs {y}')

plt.show()
```
